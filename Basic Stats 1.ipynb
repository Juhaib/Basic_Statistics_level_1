```python
import pandas as pd

# Load the dataset
df = pd.read_csv('sales_data_with_discounts.csv')
```

```python
# Select numerical columns
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Calculate mean, median, mode, and standard deviation for numerical columns
mean_vals = df[numerical_cols].mean()
median_vals = df[numerical_cols].median()
mode_vals = df[numerical_cols].mode()
std_devs = df[numerical_cols].std()

print("Mean values:\n", mean_vals)
print("\nMedian values:\n", median_vals)
print("\nMode values:\n", mode_vals)
print("\nStandard deviations:\n", std_devs)
```

```python
import matplotlib.pyplot as plt

# Plot histograms for numerical columns
df[numerical_cols].hist(bins=20, figsize=(10, 10))
plt.show()
```

```python
# Plot box plots for each numerical column
df[numerical_cols].plot(kind='box', subplots=True, layout=(2, 3), figsize=(10, 10))
plt.show()
```

```python
# Select categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Plot value counts for each categorical column as a bar plot
df[categorical_cols].apply(lambda x: x.value_counts()).plot(kind='bar', figsize=(10, 6))
plt.show()
```

```python
# Standardize numerical variables
standardized_df = (df[numerical_cols] - df[numerical_cols].mean()) / df[numerical_cols].std()

# Plot histograms of the standardized numerical variables
standardized_df.hist(bins=20, figsize=(10, 10))
plt.show()
```

```python
# Convert categorical variables into dummy/one-hot encoded variables
df_encoded = pd.get_dummies(df, columns=categorical_cols)

# Display the first few rows of the encoded dataframe
print(df_encoded.head())
```
